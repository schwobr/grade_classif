```python
import matplotlib.pyplot as plt
from pathlib import Path
import PIL.Image as Image
from math import ceil
```

```python
data = Path('/data/DeepLearning/SCHWOB_Robin/xception_weights/info')
```

```python
items = list(data.iterdir())
```

```python
ncols = min(len(items), 3)
nrows = ceil(len(items) / ncols)
_, axs = plt.subplots(nrows, ncols, figsize=(ncols*8, nrows*8))
for ax, fn in zip(axs.flatten(), items):
    img = Image.open(fn)
    ax.imshow(img)
    ax.set_title(fn.stem)
    ax.axis('off')
```
